```python
df.corr(numeric_only=True)["CRIM"].sort_values()
```

The correlations show that **crime rate (CRIM)** is positively associated with   
features like **TAX**, **LSTAT**, **NOX**, **INDUS**, and **AGE**,  
meaning areas with higher taxes, pollution, industry, and lower socioeconomic status tend to have more crime.    
Conversely, features like **DIS**, **RM**, **ZN**, and **B** are negatively correlated,  
suggesting that areas farther from the city center, with larger homes, more zoning for residential use,  
and certain demographic compositions tend to have lower crime rates.
